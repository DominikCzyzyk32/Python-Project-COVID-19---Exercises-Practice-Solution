```python
import pandas as pd

covid_data= pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/03-17-2020.csv')
```

```python
# First 5 rows
covid_data.head()
```

```python
# Data informations
covid_data.info()
```

```python
# Missing values
covid_data.isnull().sum()
```

```python
# Grouping by country
covid_data['Active'] = covid_data['Confirmed'] - covid_data['Deaths'] -  covid_data['Recovered'] 
covid_data_per_country = covid_data.groupby(["Country/Region"])[['Confirmed', 'Deaths', 'Recovered', 'Active']].apply(sum)
covid_data_per_country
```

```python
# Grouping by country and province
covid_data_per_country_and_province = covid_data.groupby(["Country/Region","Province/State"])[['Confirmed', 'Deaths', 'Recovered', 'Active']].max()
covid_data_per_country_and_province
```

```python
# Data for China
covid_data_china = covid_data.loc[covid_data['Country/Region'] == 'China']
covid_data_china = covid_data_china[['Province/State', 'Confirmed', 'Deaths', 'Recovered']].sort_values(by="Confirmed", ascending=False).reset_index(drop=True)
covid_data_china
```

```python
# Countries with the most deaths so far
covid_data_countries_with_most_deaths = covid_data.groupby(['Country/Region'])[['Deaths']].apply(sum).sort_values(by='Deaths', ascending=False)
covid_data_countries_with_most_deaths = covid_data_countries_with_most_deaths[covid_data_countries_with_most_deaths['Deaths']>0]
covid_data_countries_with_most_deaths
```
